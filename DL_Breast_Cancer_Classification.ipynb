```python
# loading the data to a data frame
import pandas as pd
from sklearn.datasets import load_breast_cancer

# Load breast cancer dataset
breast_cancer_dataset = load_breast_cancer()

# Create DataFrame from the dataset
data = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)

# Print the first few rows of the DataFrame to verify
```

```python
# Adding the 'target' column to the dataframe
data['label']=breast_cancer_dataset.target
```

```python
# Separating the Features and target
X=data.drop('label',axis=1)
Y=data['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train_std=scaler.fit_transform(X_train)
X_test_std=scaler.fit_transform(X_test)
```

```python
print(X_train_std)
```

# Building the Neural Network

```python
# Importing tenserflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
import keras
from keras.layers import Flatten, Dense
```

```python
# Setting up the layer of Neural Network
model = keras.Sequential([
    Flatten(input_shape=(30,)),  # Specify input shape here
    Dense(20, activation='relu'),
    Dense(2, activation='sigmoid')
])
```

```python
# compiling the Neural Network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Training the Nerual Network
history=model.fit(X_train_std,Y_train,validation_split=0.1,epochs=10)
```

```python
# Visualizing accuracy and loss
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['training_data','validation data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epochs')

plt.legend(['training data','validation data'],loc='upper right')
```

```python
loss,accuracy=model.evaluate(X_test_std,Y_test)
print(accuracy)
```

```python
Y_pred=model.predict(X_test_std)
```

```python
print(Y_pred.shape)
print(Y_pred[0])
```

```python
# Converting the prediction probabulity to class labels
Y_pred_labels=[np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```
